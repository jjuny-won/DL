# 학습 관련 기술들

## 6.1 매개 변수 갱신
- 최적화 : 손실 함수의 값을 가능한 한 낮추는 매개변수를 찾는 것 -> 매개변수의 최적값을 찾는 문제


### 6.1.2 확률적 경사 하강법(SGD)


- 학습률 : 0.01 또는 0.001 과 같은 값을 정해서 사용
- 기울어진 방향으로 일정 거리만 가겠다는 단순 한 방법

#### 단점
- 비등방성 함수(기울기가 달라지는 함수)에서는 탐색 경로가 비효율적
- 무작정 기울어진 방향으로 진행하는 단순한 방식보다 영리한 묘얀이 필요!

### 6.1.4 모멘텀

- W는 가중치 매개변수, v : 물리에서 말하는 속도
- 기울기 방향으로 힘을 받아 물체가 가속된다는 물리 법칙을 나타낸다
- av : 아무런 힘을 받지 않을 때 서서희 하강시키는 역할 (0.9)
- 전체적으로 SGD 보다 x 축 방향으로 빠르게 다가가 지그재그로 움직임이 줄어든다

### 6.1.5 AdaGrad
- 학습률 감소 ( learning rate dacay ) : 학습을 진행하면서 진행률을 점차 줄여나가는 방법
- '갹갹의' 매개변수에 '맞춤형' 값을 만들어준다
- 개별 매개변수에 적응적으로 학습률을 조정하면서 학습 진행
- 과거의 기울기를 제곱하여 계속 더해가는 방식 -> 무한히 학습한다면 어느 순간 갱신량이 0이 되어버림
- RMSProp ( 먼 과거의 기울기는 잊고 새로운 기울기 정보 크게 반영 , 지수이동평균 )


### 6.1.6 Adam
- 모멘텀 + AdaGrad
- 하이퍼파라미터의 '편향 보정'이 진행된다





## 6.2 가중치의 초깃값

### 6.2.1 초깃값을 0으로 하면?
- 가중치 감소 ( weight decay ) : 가중치 매개변수의 값이 작아지도록 학습하는 방법
- 모둔 균일하게 0으로 설정할 경우 - 모든 가중치의 값이 똑같이 갱신되기 떄문

### 6.2.2 은닉층의 활성화값 분포
- 가중치의 초깃값에 따른 은닉층 활성화 값들의 변화
- 각 층이ㅡ 활성화 값은 적당히 고루 분포해야한다
- 층과 층 사이에 다양한 데이터가 흐르게 해야 신경망 학습이 효율적으로 이뤄지기 떄문
- 치우친 데이터가 흐르면 기울기 소실 문제, 표현력 제한 문제에 빠져 학습이 잘 이뤄지지 않는다

#### Xavier 초깃값
- 전제 : 활성화 함수가 선형일 때
#### ReLU 를 사용할 때 가중치 초깃값
- He 초깃값 : Xavier 보다 2배의 계수가 필요

- 궁금한점 ? Xavier Normal , Uniform 의 차이점?





## 6.3 배치 정규화
 
 
### 6.3.1 배치 정규화 알고리즘
- 각 층에서 활성화 값이 적당히 분포되도록 조정하는 것
- 학습 시 미니배치를 단위로 정규화 시킨다 (평균이 0, 분산이 1)
- 미니 배치 입력데이터를 평균 0, 분산 1인 데이터로 변환시켜  활성화 함수 앞에 삽입해 분포가 치우치는 것을 막는다
- 배치 정규화 계층마다 이 정규화된 데이터에 고유한 확대, 이동 변환 수행

#### 배치 정규화를 사용하는 이유
- 학습 진도가 빠르다
- 가중치 초깃값에 크게 의존하지 않아도 된다




## 6.4 오버피팅
- 신경망이 훈련 데이터에만 지나치게 적응되어 그 외의 데이터에는 제대로 대응하지 못하는 상태

### 6.4.1 오버 피팅
- 매개 변수가 많고 표현력이 높은 모델
- 훈련 데이터가 적을 때

### 6.4.2 가중치 감소
- 학습 과정에서 큰 가중치에 대해서는 그에 상응하는 큰 패널티를 부과하여 오버피팅을 억제하는 방법
- 신경망의 목적은 손실 함수의 값을 줄이는 것. 갖ㅇ치의 제곱 노름을 더한다면 가중치가 커지는 것을 억제할 수 있다
- L2 노름 : 각 원소들의 제곱들을 더한 것
- L1 노름 : 절댓값의 합

### 6.4.3 드롭 아웃
- 뉴런을 임의로 삭제하면서 학습하는 방법
- 훈련 때 은닉층의 뉴런을 무작이로 골라 삭제하여 신호 전달 차단
- 표현력을 높이면서 오버 피팅을 억제할 수 있다 

- ** 앙상블 학습 **
- 개별적으로 학습시킨 여러모델의 출력을 평균내어 추론하는 방식
- 매번 다른 모델을 학습시키는 것 -> 뉴런의 출력에 삭제한 비율을 곱함으로써 여러 모델의 평균을 내는 것 (드롭아웃과 같은 맥락)

## 6.5 적절한 파라미터 값 찾기

### 6.5.1 검증 데이터
- 하이퍼파라미터의 성능을 평가할 때는 시험 데이터를 사용해서는 안된다는 것!
- 검증 데이터 : 하이퍼파라미터 전용 확인 데이터

#### 훈련 데이터 : 매개변수 학습
#### 검증 데이터 : 하이퍼파라미터 성능 평가
#### 시험 데이터 : 신경망의 범용 성능 평가


### 6.5.2 하이퍼파라미터 최적화
- 하이퍼파라미터의 '최적 값'이 존재하는 범위를 조금씩 줄여나가는 것

#### 0 단계 : 하이퍼파라미터의 값의 범위를 설정
#### 1단계 : 설정된 범위에서 하이퍼파리미터의 값을 무작위로 추출
#### 2단계 : 1단계에서 샘플링한 하이퍼파라미터의 값을 사용해 학습하고, 검증 데이터로 정확도 평가(에폭은 작게)
#### 3단계 : 1단계, 2단계 특정 횟수 반복해, 정확도의 결과를 보고 하이퍼파라미터의 범위를 좁혀나감
